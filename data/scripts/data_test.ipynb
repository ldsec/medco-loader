```python
import pandas as pd

clinical_data = pd.read_csv("../genomic/tcga_cbio/clinical_data.csv", sep='\t', dtype=object)
mutation_data = pd.read_csv("../genomic/tcga_cbio/mutation_data.csv", sep='\t', dtype=object)
```

```python
group_variants = mutation_data.groupby('START_POSITION')
sr = group_variants.apply(lambda x: len(x['PATIENT_ID'].unique()))
#nbr_patients_per_variant
nppv = pd.DataFrame({'START_POSITION':sr.index, 'count':sr.values})
```

```python
df_position = nppv.loc[nppv['count']==15]
pos = df_position.iloc[0]['START_POSITION']

df = mutation_data.loc[mutation_data['START_POSITION']==pos]
```

```python
data_query = df.head(1)
patient_ids = df['PATIENT_ID']
for _, val in patient_ids.iteritems():
    tmp = (mutation_data.loc[mutation_data['PATIENT_ID']==val]).head(1)
    data_query = data_query.append(tmp)
    
other_variants = mutation_data.head(200-16)
data_query = data_query.append(other_variants)

export_csv = data_query.to_csv(r'query.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
data_query
```

```python
and_query = data_query.head(1)
or_query = data_query.iloc[1:]

sp_and = and_query['START_POSITION']
ep_and = and_query['END_POSITION']
tsa1_and = and_query['TUMOR_SEQ_ALLELE1']
tsa2_and = and_query['TUMOR_SEQ_ALLELE2']

sp = or_query['START_POSITION']
ep = or_query['END_POSITION']
tsa1 = or_query['TUMOR_SEQ_ALLELE1']
tsa2 = or_query['TUMOR_SEQ_ALLELE2']

query_part1 = mutation_data.loc[mutation_data['START_POSITION'].isin(sp) & mutation_data['END_POSITION'].isin(ep) & mutation_data['TUMOR_SEQ_ALLELE1'].isin(tsa1) & mutation_data['TUMOR_SEQ_ALLELE2'].isin(tsa2)]
pIDs = query_part1['PATIENT_ID'].values
patients_query_part1 = clinical_data.loc[clinical_data['PATIENT_ID'].isin(pIDs)]

data2 = mutation_data.loc[mutation_data['PATIENT_ID'].isin(pIDs)]
query_part2 = data2[data2['START_POSITION']==sp_and.values[0]]
query_part2 = query_part2[query_part2['END_POSITION']==ep_and.values[0]]
query_part2 = query_part2[query_part2['TUMOR_SEQ_ALLELE1']==tsa1_and.values[0]]
query_part2 = query_part2[query_part2['TUMOR_SEQ_ALLELE2']==tsa2_and.values[0]]
query_part2
```

```python
short_data = mutation_data.drop(columns=['CANCER_STUDY_ID','CANCER_STUDY_IDENTIFIER','SAMPLE_ID','TUMOR_SEQ_ALLELE1','TUMOR_ALT_COUNT','TUMOR_REF_COUNT','NORMAL_ALT_COUNT','NORMAL_REF_COUNT','PROTEIN_CHANGE','MUTATION_TYPE','NCBI_BUILD','STRAND','VARIANT_TYPE','ENTREZ_GENE_ID','HUGO_GENE_SYMBOL'])
short_data.to_csv(r'mutation_data_small.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
```
